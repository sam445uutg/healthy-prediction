```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.layers import Dense
from keras.models import Model
import matplotlib.pyplot as plt
```

```python
food_coded = pd.read_csv('/content/drive/MyDrive/food_coded.csv')
#print(food_coded.dtypes)
u_l=[]
for col_name in food_coded:
  if(food_coded[col_name].dtype == 'object'):
    u_l.append(col_name)

print(u_l)
```

```python
data_sets = food_coded.drop(u_l,axis='columns')
data_sets.fillna(0, inplace=True)
data_sets
```

```python
#print(data_sets.isnull().sum().sort_values(ascending=False))
data_sets.mean().sum()
```

```python
data_sets.insert(49, 'health', ((data_sets.weight <=155) | (data_sets.weight >=140)) )
```

```python
#, input_spec
from keras.engine import input_layer, input_spec
from keras.engine.input_layer import Input
import tensorflow as tf
from keras.layers import Dense, Lambda 
from keras.models import Model, Sequential

def preprocess(x):
  nomraliztion=tf.keras.layers.Normalization()
  return nomraliztion(x)
def layer(x):
  layers1= tf.keras.layers.Dense(250,activation='relu')(x)
  layers2= tf.keras.layers.Dense(1,activation='sigmoid')(layers1)
  return layers2
inputs= tf.keras.Input(shape=(48,))
pre= preprocess(inputs)
outputs=layer(pre)
model=tf.keras.Model(pre, outputs)
'''model= Sequential([
                   tf.keras.layers.Dense(96, input_shape=(48,), activation='relu'),
                   tf.keras.layers.Dense(192, activation='relu'),
                   tf.keras.layers.Dense(96,activation='relu'),
                   tf.keras.layers.Dense(1,)
])'''
```

```python

model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
#train_datads
from sklearn.model_selection import train_test_split
x=data_sets.drop(['weight','health'],axis='columns')
y=data_sets['health']
train_x, test_x,  train_y, test_y =train_test_split(x,y, test_size=0.1, random_state=50)
```

```python
model.fit(train_x, train_y, batch_size=48, epochs=100, validation_data=(test_x, test_y))
```

```python
history =model.evaluate(test_x, test_y)
```
